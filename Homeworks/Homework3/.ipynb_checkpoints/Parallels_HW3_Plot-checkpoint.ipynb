```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Get_Data with P = 3
sorted_data = pd.read_csv("Result_P3_N10.txt", sep="/", header=None)
np.asarray(sorted_data)
time = np.arange(sorted_data.size)
plt.scatter(time,sorted_data,s = 4,c = 'blue')
plt.title("Sorted data (P =3)")
plt.xlabel('N = 10')
plt.show()
```
